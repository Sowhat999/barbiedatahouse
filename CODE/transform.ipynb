```python
import pandas as pd

df_scr = pd.read_csv('DATA/careers_scraped.csv')
df = pd.read_csv('DATA/barbie_careers.csv')
```

```python
missing_years = df[df['Year'].isna()]
missing_years
```

```python
#Since only 4 years are missing we can look them up and add manually
#create dictionary of misssing barbies and years
career_year = {'Special education teacher' :2020, 'Geologist': 2021, 'Photographer': 2014, 'Banker': 2019} 

for key, value in career_year.items():
    df.loc[df['Career'] == key, 'Year'] = value
```

```python
df['Type'].value_counts()
```

```python
df_scr['Type']= None
```

```python
#need to assign type of employment to the scraped careers data set
#need to change some types of careers, extra career fields are below
retail_customer = ["retail_customer","Resort greeter","McDonald's cashier", "Soda founta waitress (Coca Cola edition)","Waitress","See's Candies cashier"]
restaurant_kitchen = ["restaurant_kitchen","Noodle bar worker","Coffee shop worker", "Chef"]
farming = ['farming', 'Bee Keeper', 'Chicken farmer']
education = ['Education', 'Teacher','Nursery School', 'Baby Sitter','Babysitter']
sport = ['Sport', 'Player', 'Gymnast']
medicine = ['Medicine', 'Nurse','doctor','Dentist']
transportation = ['Transportation', 'Attendant','Stewardess','Lifeguard','Travel Agent']
public = ['Public Service','President','Firefighter','Secretary']
business = ['Business','Entrepreneur']
arts = ['Arts','Magician','Photographer','Star']
science = ['Science and engineering','Engineer','Explorer']
animal_care = ['animal_care','Zookeeper','Dolphin Trainer','Pet groomer']
media = ['Media', 'News']
new_categories = [retail_customer,restaurant_kitchen,farming, medicine, education, sport, transportation, public, business, arts, science,animal_care, media]
```

```python
for category in new_categories:
    filter = df_scr['Career'].apply(lambda x: any(substring.lower() in x.lower() for substring in category))
    df_scr.loc[filter, 'Type'] = category[0]
```
