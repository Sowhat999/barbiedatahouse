```python


# Get the number of unique titles
num_unique_titles = barbiedf['Title'].nunique()
print(f"Number of unique titles: {num_unique_titles}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
from scipy.stats import ks_2samp
from scipy import stats


sns.histplot(barbiedf['Highest'],kde= True, color ='#e0218a')
sns.histplot(barbiedf['Lowest'], kde= True, color = '#ff6fff' )
plt.legend(loc= 'upper right', labels = ['Highest Barbie Cost','Lowest Barbie Cost'])
plt.title('Highest vs Lowest Barbie', fontsize=12, fontweight='normal') 
```

```python



# Group data by title and calculate the count of each title
titlecounts = barbiedf['Title'].value_counts()

# Get the top 10 titles
top_10_titles = titlecounts.head(20)


# Create a bar chart with pink bars for the top 10 titles
plt.figure(figsize=(10, 6))
ax=top_10_titles.plot(kind='bar', color='#e0218a')
plt.title('Top 20 Collectible Barbies')
plt.xlabel('Barbies')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
ax.yaxis.set_major_locator(plt.MaxNLocator(integer=True))

plt.tight_layout()  
plt.tight_layout()

# Save the plot to a file
#plt.savefig('pink_top_10_titles.png')

# Show the plot
plt.show()

```

```python

releasecounts = barbiedf['Released'].value_counts().sort_index()

# Create a bar plot for the counts of releases by year
plt.figure(figsize=(10, 6))
releasecounts.plot(kind='bar', color='#e0218a') 
plt.title('Barbie Releases by Year')
plt.xlabel('Year')
plt.ylabel('Number of Releases')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()


#plt.savefig('barbie_releases_by_year.png')

# Show the plot
plt.show()
```

```python


# Select the columns for the scatter plot
xcolumn = 'Highest'
ycolumn = 'Lowest'

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(barbiedf[xcolumn], barbiedf[ycolumn], color='#e0218a', alpha=0.5)
plt.title(f'Scatter Plot of {xcolumn} vs {ycolumn}')
plt.xlabel(xcolumn)
plt.ylabel(ycolumn)
plt.grid(True)
plt.tight_layout()


#plt.savefig('barbie_scatter_plot.png')

# Show the plot
plt.show()
```
