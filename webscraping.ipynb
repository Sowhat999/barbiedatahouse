```python
# Importing libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
from IPython.display import Image, HTML
```

```python
titles = []
product_numbers = []
releases = []
lowest_prices = []
average_prices = []
highest_prices = []
barbie_url = []
```

```python
url = 'https://www.barbiedb.com/'

response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
    # Now, you can use Beautiful Soup methods to extract data from the page
    # For example, let's print the titles of all the articles on the page:
    for article in soup.find_all('article'):
        title = article.find('h2').text.strip()
        titles.append(title)
        release_year = article.find('label', class_='barbie__year').input['value']
        releases.append(release_year)
        product_number_final = article.find('label',class_='barbie__product_number').input['value']
        product_numbers.append(product_number_final)
        lowest_price = article.find('label', class_='barbie__value_lowest').input['value']
        lowest_prices.append(lowest_price)
        highest_price = article.find('label', class_='barbie__value_highest').input['value']
        highest_prices.append(highest_price)
        average_price = article.find('label', class_='barbie__value_average').input['value']
        average_prices.append(average_price)
        image_link = article.find('img',itemprop='image')
        image_url = image_link['data-src']
        barbie_url.append(image_url)
        print(product_number_final)
        print(title)
        print(average_price)
        print(highest_price)
        print(lowest_price)
        print(image_url)
       
        
else:
    print(f"Failed to fetch the page. Status code: {response.status_code}")
```
